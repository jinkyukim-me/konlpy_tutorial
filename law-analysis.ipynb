```python
import nltk
```

KoNLPy의 내부 문서 중 육아휴직 관련 법안 제 1809890호를 읽습니다. 

이 법률은 비록 공무원에 한정하고 있지만 육아휴직 대상이 되는 아이의 나이를 만 6세에서 만 8세로 확장하는 내용을 담고 있습니다.

```python
from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko
```

```python
from konlpy.tag import Okt 
t = Okt()
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
```

```python
print(len(ko.tokens))           # returns number of tokens (document length) 수집한 단어의 횟수
print(len(set(ko.tokens)))   # returns number of unique tokens 고유한 횟수
ko.vocab()                        # returns frequency distribution
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번',
                      '호','을','이','다','만','로','가','를']

ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')

plt.figure(figsize=(12,6))
ko.plot(50)     # Plot sorted frequency of top 50 tokens
plt.show()
```
